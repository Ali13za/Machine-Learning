## Array creation
>Create the followings to numpy array:
* A: a 3\*3 array like this:
$$
A = \left(\begin{array}{cc} 
10 & 20 & 30\\
5 & 10 & 15\\
1 & 5 & 8
\end{array}\right)
$$ 
* B: a 3\*3\*4 that filled with the numbers start from 15 and have steps of 5 to the end. The array should be filled first along the first dimension then the second dimension and at last the third dimension. like this:
$$
B[:, :, 0] = \left(\begin{array}{cc} 
15 & 30 & 45\\
20 & 35 & 50\\
25 & 40 & 55
\end{array}\right)
$$ 

```python
### START CODE HERE ###
A = np.array([[10, 20, 30], [5, 10, 15], [1, 5, 8]])

B = np.array([[[15, 60, 105, 150], [30, 75, 120, 165], [45,  90, 135, 180]],
              [[20, 65, 110, 155], [35, 80, 125, 170], [50,  95, 140, 185]],
              [[25, 70, 115, 160], [40, 85, 130, 175], [55, 100, 145, 190]]])
print(B)
### END CODE HERE ###
```

## Replacement
>Replace the last slice of array B (B[:, :, 3]) with values of A. Then print the result.

```python
### START CODE HERE ###
B[:,:,3] = A
print(B)
### END CODE HERE ###
```

## Axis Swap 
>Swap the first(B[:, :, 0]) and last(B[:, :, 3]) slice of the array B **only with one line of code**. Then print the result.

```python
### START CODE HERE ###
B[:,:,[0,3]] = B[:,:,[3,0]]
print(B)
### END CODE HERE ###
```

## Transposing 
>Transpose the array B elements in each slice. Then print the result.

```python
### START CODE HERE ###
for i in range(4):
    B[:,:,i] = np.transpose(B[:,:,i])
print(B)
### END CODE HERE ###
```

## Matrix Multiplication
>Calculate the matrix multiplication of all slices of the array B. 
(B[:, :, 0] \* B[:, :, 1] \* B[:, :, 2] \* B[:, :, 3]). Then print the result.

>**Note**: the result should be a 3\*3 array. 

```python
### START CODE HERE ###
M = np.matmul(np.matmul(B[:,:,0],B[:,:,1]),np.matmul(B[:,:,2],B[:,:,3]))
print(M)
### END CODE HERE ###
```

## Loading dataset
>Load the IMDB-Movie-Data.csv file into a pandas dataframe and show its first 5 rows. Each row of this dataframe corresponds to one film.

```python
### START CODE HERE ###
data = pd.read_csv('IMDB-Movie-Data.csv')
data.head()
### END CODE HERE ###
```

## Actors/Actresses Activity
>For each Actors/Actresses calculate the number of movies in the dataframe, then sort the actors/actresses first by their number of movies in descending order and then by their name in ascending order. 

```python
### START CODE HERE ###
actors_activity = {}
for actors in data['Actors']:
    for actor in actors.split(','):
        if actor.strip() in actors_activity:
            actors_activity[actor.strip()] += 1
        else:    
            actors_activity[actor.strip()] = 1
sorted_actors_activity = sorted(actors_activity.items(), key=lambda x:(-x[1],x[0]))
sorted_actors_activity
### END CODE HERE ###
```

## Score Histogram

## Plot
>You must plot the cos(x) curve. plot 20 cycle of the curve. the frequency of first half of curve is 50 Hz and frequency of second lalf is 200 Hz. 
> set the x_label, y_label and title for your plot.

```python
### START CODE HERE ###
f1, f2 = 50, 200
T1, T2 = 10/f1, 10/f2
x = np.linspace(0, T1+T2, 1000)
y = np.cos(2*np.pi*((x<T1)*f1 + (x>T1)*f2)*x)
plt.plot(x,y)
plt.title('ten 50Hz cycles & ten 200Hz cycles')
plt.xlabel('time')
plt.ylabel('cos(x)')
plt.show()
### END CODE HERE ###
```
